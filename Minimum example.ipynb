```python
import numpy as np
import pandas as pd
import xarray as xr
import os
import cartopy.crs as ccrs

import matplotlib.pyplot as plt
%matplotlib inline
```

## Minimal example

```python
with xr.open_dataset('snow_rlat_rlon.nc') as ds:
    snow = ds.snowfall
    
snow *= 86400 * 365 # mm/s to mm/yr
```

```python
#levels = np.linspace(0, 2000, 21)
levels = (0, 25, 50, 75, 100, 200, 400, 600, 800, 1000, 2000, 4000)
snow.plot.contourf(levels=levels)

plt.title('Contourf (no projection)')
```

```python
plt.figure()
ax = plt.subplot(111,projection=ccrs.SouthPolarStereo())

snow.plot.pcolormesh('lon2d', 'lat2d', ax=ax, 
                                transform=ccrs.PlateCarree(), add_colorbar=False, levels=levels, 
                                extend='both')

ax.coastlines()
plt.title('Pcolormesh (projected)')
```

```python
plt.figure()
ax = plt.subplot(111,projection=ccrs.SouthPolarStereo())

snow.plot.contourf('lon2d', 'lat2d', ax=ax, 
                                transform=ccrs.PlateCarree(), add_colorbar=False, levels=levels, 
                                extend='both')
ax.coastlines()
plt.title('contourf (projected)')
```
